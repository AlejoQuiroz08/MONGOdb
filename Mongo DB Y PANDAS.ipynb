```python
from pymongo import MongoClient

# Conéctate a la base de datos MongoDB
client = MongoClient('mongodb://localhost:27017/')
database = client['database1']
collection = database['coleccion1']

# Insertar datos
data_to_insert = {"nombre": "Ejemplo", "edad": 25}
insert_result = collection.insert_one(data_to_insert)
print(f"ID del documento insertado: {insert_result.inserted_id}")

# Leer datos
query = {"nombre": "Ejemplo"}
result = collection.find_one(query)
print(f"Dato encontrado: {result}")

# Actualizar datos
update_query = {"nombre": "Ejemplo"}
update_data = {"$set": {"edad": 26}}
update_result = collection.update_one(update_query, update_data)
print(f"Número de documentos modificados: {update_result.modified_count}")

# Eliminar datos
delete_query = {"nombre": "Ejemplo"}
delete_result = collection.delete_one(delete_query)
print(f"Número de documentos eliminados: {delete_result.deleted_count}")


```

```python

from pymongo import MongoClient
import pandas as pd
client=MongoClient('127.0.0.1', 27017)
database= client['BaseNetflix']
collection= database['IMDB']
cursor= collection.find()
listado= list(cursor)
df1= pd.DataFrame(listado)

client.close()

print("Datos de la pelicula Fine Wine")
df1[(df1["title"]=="Fine Wine")]



```

```python
print("Shows estrenados despues de 2020")
df1[(df1["type"]=="SHOW")&(df1["release_year"]>2020)]
```

```python
print("Shows con mas calificacion que 8")
df1[(df1["type"]=="SHOW")&(df1["imdb_score"]>8)]
```

```python
print("Las 10 peliculas mayor clasificadas")
df1[(df1["type"]=="MOVIE")&(df1["imdb_score"]>8)].head(10)


```
